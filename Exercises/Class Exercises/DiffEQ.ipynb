```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

**Section 3, Exercise 1**

```python
def f(x,t):
    return math.cos(t)

a = 0.0 # Start of the interval
b = 1 # End of the interval
N = 1000 # Number of steps

h1 = 1.0 # Size of step
h2 = 0.5
x_1 = 0.8415 # Initial

X_h1 = [x_1]
X_h2 = [x_1]
T_h1 = [1]
T_h2 = [1]

for i in range(10):
    X_h1.append(X_h1[i] + h1*f(X_h1[i], t=T_h1[i]))
    T_h1.append(T_h1[i] + h1)
    X_h2.append(X_h2[i] + h2*f(X_h2[i], t=T_h2[i]))
    T_h2.append(T_h2[i] + h2)

```

```python
print(X_h1)
print(X_h2)
```

Fourth Order Runge-Kutta Method:

k1 = hf(x,t) |--|
k2 = hf(x+k1/2, t+h/2) |--|
k3 = hf(x+k2/2, t+h/2) |--|
k4 = hf(x+k3, t+h) |--|
x(t+h) = x(t) + 1/6(k1 + 2k2 + 2k3 + k4)

**Section 3, Exercise 2**

```python
def f(x, t):
    return math.exp(-2*t)

def g(x, u):
    return 1/(1-u)**2*f(x, u/(1-u))

a = 0
b = 1
N = 1000
h = (b-a)/N
x = 10

U = np.arange(a, b, h)
T = []
X = []

for u in U:
    X.append(x)
    x += h*g(x, u)
    t = u/(1-u)
    T.append(t)


plt.plot(T, X)
plt.xlabel('t')
plt.xlim(0,10)
plt.ylabel('x')
plt.title('t vs x')
plt.savefig('T vs X.png', dpi=300)
plt.show()
```
