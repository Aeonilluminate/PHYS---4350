### Section 5, Exercise 1

Write a program to simulate a coin toss by randomly generating 1's (heads) or 0's (tails). Generate 100 numbers and count how many times the coin comes up heads.

```python
import random
import matplotlib.pyplot as plt
```

```python
def flip():
    return random.randint(0, 1)

result_dict = {1: 'H', 0: 'T'}

flips = 0
heads = 0
results = []
while flips < 100:
    result = flip()
    heads += result
    results.append(result_dict[result])
    flips += 1

print(heads)
```
